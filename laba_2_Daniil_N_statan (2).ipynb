## –¶–µ–ª—å —Ä–∞–±–æ—Ç—ã
–ù–∞—É—á–∏—Ç—å—Å—è –ø—Ä–∏–º–µ–Ω—è—Ç—å –º–µ—Ç–æ–¥—ã –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞; —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å –Ω–∞–≤—ã–∫–∏ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –º–µ—Ç–æ–¥–æ–≤ —Å –ø–æ–º–æ—â—å—é –ø—Ä–∏–∫–ª–∞–¥–Ω—ã—Ö –ø—Ä–æ–≥—Ä–∞–º–º–Ω—ã—Ö –ø–∞–∫–µ—Ç–æ–≤.



## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è

$–ú–∞—Ç.–æ–∂–∏–¥–∞–Ω–∏–µ$<br>
$Ex=\frac{{\displaystyle\sum_{i=1}^n}x_i}n$<br>
$–î–∏—Å–ø–µ—Ä—Å–∏—è:$<br>
$\operatorname{ùê∑}X=E\left(X-E\left(X\right)\right)^2$<br>
$–°–ö–û:$<br>
$s=\left(\operatorname{ùê∑}X\right)^\frac12$<br>
$–ê—Å—Å–∏–º–µ—Ç—Ä–∏—è:\\\widehat{Skew}\left(x\right)=\frac{{\displaystyle\frac1n}{\displaystyle\sum_{i=1}^n}\left(x_i-\overline x\right)^3}{s^3}$<br>
$–≠–∫—Å—Ü–µ—Å—Å:\\\widehat{Kurt}\left(x\right)=\frac{\displaystyle\frac1n\sum_{i=1}^n\left(x_i-\overline x\right)^4}{s^4}-3$<br>
$–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å\;–ø–æ–ø–∞–¥–∞–Ω–∏—è:\\P\left(x\in\left[a,b\right]\right)=F\left(b\right)-F\left(a\right)$<br>
$–ü–ª–æ—Ç–Ω–æ—Å—Ç—å\;—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è\;–ü—É–∞—Å—Å–æ–Ω–∞:$<br>
$P_\theta\left(x\right)=\frac{\lambda^x}{x!}e^{-\lambda}$<br>
$–ü–ª–æ—Ç–Ω–æ—Å—Ç—å\;–Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ\;—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:$<br>
$P_\theta\left(x\right)=\frac1{\sigma\sqrt{2\pi}}exp\left(-\frac{\left(x-\mu\right)^2}{2\sigma^2}\right)$<br>

$–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞\;–∫—Ä–∏—Ç–µ—Ä–∏—è\;–ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞:$<br>
$D_n=sup_{x\in\mathbb{R}}\left|F_n\left(x\right)-F_0\left(x\right)\right|$<br>
$–¢–µ–æ—Ä–µ–º–∞\;–ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞:$<br>
$\sqrt nD_n\Rightarrow\;K\\P_{F_0}\left(\sqrt nD_n<x\right)\rightarrow K\left(x\right),\mathbb{R}\left(x\right)$<br>
$–≥–¥–µ\;K\;\;—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ\;–ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞$<br>
$–ì–∏–ø–æ—Ç–µ–∑–∞\;–æ—Ç–∫–ª–æ–Ω—è–µ—Ç—Å—è,\;–µ—Å–ª–∏\;$<br>
$D_n>x$<br>
$–ü—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è,\;–µ—Å–ª–∏\;$<br>
$D_n<x$<br>
$–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞\;–æ—Ç–Ω–æ—à–µ–Ω–∏—è\;–ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è:$<br>
$LR\left(x\right)=\frac{L\left(x,\theta_1\right)}{L\left(x,\theta_0\right)}=\left(\frac{\theta_1\left(1-\theta_0\right)}{\theta_0\left(1-\theta_1\right)}\right)^{\sum_{i=1}^nx_i},\\–≥–¥–µ\;L\left(x,\theta_1\right)\;–∏\;L\left(x,\theta_0\right)\;-\;—Ñ—É–Ω–∫—Ü–∏–∏\;–ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è$<br>
$–ó–∞–ø–∏—Å—å\;–∫—Ä–∏—Ç–µ—Ä–∏—è\;–ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞\;–∏\;–∑–∞–ø–∏—Å—å\;–≤\;—Ç–µ—Ä–º–∏–Ω–∞—Ö\;\overline X:$<br>
$–§\left(x\right)=\left\{\begin{array}{l}1,\;LR\left(x\right)>C\\p,\;LR\left(x\right)=C\\0,\;LR\left(x\right)<C\end{array}\right.$<br>
$–§\left(x\right)=\left\{\begin{array}{l}1,\;\overline X<C^\ast\\p,\;\overline X=C^\ast\\0,\;\overline X>C^\ast\end{array}\right.\\LR\left(X\right)>C\Leftrightarrow\overline X<C^\ast\\LR\left(X\right)<C\Leftrightarrow\overline X>C^\ast$<br>

$–û–ú–ü:$<br>
$–û—Ü–µ–Ω–∫–∞\;\widehat\theta\left(x\right),\;–º–∞–∫—Å–∏–º–∏–∑–∏—Ä—É—é—â–∞—è\;—Ñ—É–Ω–∫—Ü–∏—é\;–ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è\;L\left(x;\theta\right)\;–ø–æ\;\theta\;–ø—Ä–∏\;–∫–∞–∂–¥–æ–º\\—Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º\;–•:$<br>
$L\left(X;\widehat\theta\left(X\right)\right)\geq L\left(X;\widetilde\theta\left(X\right)\right),\;\forall X\in\mathfrak X,\;\forall\widetilde\theta\left(X\right)$<br>
$–Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è\;–æ—Ü–µ–Ω–∫–æ–π\;–º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ\;–ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è$<br>
$–û—Ü–µ–Ω–∫–∞\;—è–≤–ª—è–µ—Ç—Å—è\;–Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–π,\;–µ—Å–ª–∏\;–≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è\;—Å–ª–µ–¥—É—é—â–µ–µ\;—É—Å–ª–æ–≤–∏–µ:$<br>
$E\left(\widehat\theta\left(x\right)\right)=\theta$<br>
$–û–ú–ú:$<br>
$–ï—Å–ª–∏\;—Å—É—â–µ—Å—Ç–≤—É–µ—Ç\;–µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ\;—Ä–µ—à–µ–Ω–∏–µ\;\widetilde\theta\left(x\right)\;—Å–∏—Å—Ç–µ–º—ã\;—É—Ä–∞–≤–Ω–µ–Ω–∏–π:$<br>
$\widehat{\mu_k}\left(\widetilde\theta\right)=\mu_k\left(\widetilde\theta\right),\;k=1,2..d\\–≥–¥–µ\;\widehat{\mu_k}\left(\widetilde\theta\right)=\frac1n\sum_{i=1}^nx_i^k\;-\;–≤—ã–±–æ—Ä–æ—á–Ω—ã–π\;–º–æ–º–µ–Ω—Ç\;k-–æ–≥–æ\;–ø–æ—Ä—è–¥–∫–∞,\;—Ç–æ\;\widetilde\theta\left(x\right)$<br>
$–Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è\;–æ—Ü–µ–Ω–∫–æ–π\;–ø–æ\;–º–µ—Ç–æ–¥—É\;–º–æ–º–µ–Ω—Ç–æ–≤$<br>
$–°–º–µ—â–µ–Ω–∏–µ–º\;–Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è\;–≤–µ–ª–∏—á–∏–Ω–∞:\\bias\widehat{{}_\theta}\left(\theta\right)=E_{{}_\theta}\widehat{{}_\theta}\left(x\right)-\theta$<br>

$–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π\;–∏–Ω—Ç–µ—Ä–≤–∞–ª:$<br>
$–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞\;\widehat\theta:\mathfrak X\rightarrow\mathbb{R}^2,\;—É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∞—è\;—É—Å–ª–æ–≤–∏—é\;$<br>
${\mathbb{P}}_\theta\left(\left[–¢_1\left(\overrightarrow x\right),–¢_2\left(\overrightarrow x\right)\right]\ni\theta\right)\geq1-\alpha$<br>
$–Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è\;–¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–º\;–∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–º\;–ø–∞—Ä–∞–º–µ—Ç—Ä–∞\;\theta\;—É—Ä–æ–≤–Ω—è\;–¥–æ–≤–µ—Ä–∏—è\;1-\alpha$<br>

$–ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∏–π\;–¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π\;–∏–Ω—Ç–µ—Ä–≤–∞–ª:$<br>
$–ù–∞\;–±–∞–∑–µ\;–û–ú–ü$<br>
$\left[\widehat\theta-\frac{x_\alpha}{\sqrt{nl\left(\widehat\theta\right)}};\widehat\theta+\frac{x_\alpha}{\sqrt{nl\left(\widehat\theta\right)}}\right]\ni\theta$<br>
$–ù–∞\;–±–∞–∑–µ\;–∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∏\;–Ω–æ—Ä–º–∞–ª—å–Ω–æ–π\;–æ—Ü–µ–Ω–∫–∏\;–ø–∞—Ä–∞–º–µ—Ç—Ä–∞\;\widehat\theta:$<br>
$\left[\widehat\theta\left(\overrightarrow x\right)-\frac{x_\alpha\times\sigma\left(\widehat\theta\left(\overrightarrow x\right)\right)}{\sqrt n};\widehat\theta\left(\overrightarrow x\right)+\frac{x_\alpha\times\sigma\left(\widehat\theta\left(\overrightarrow x\right)\right)}{\sqrt n}\right]\ni\theta$<br>

$–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π\;–≥–∏–ø–æ—Ç–µ–∑–æ–π\;–Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è\;—É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ\;–æ\;–ø–∞—Ä–∞–º–µ—Ç—Ä–µ\;–≤–∏–¥–∞:$<br>$H:\theta\in\Theta^\ast\subseteq\;\Theta$<br>
$–û–¥–Ω–æ—Ç–æ—á–µ—á–Ω–∞—è\;–≥–∏–ø–æ—Ç–µ–∑–∞\;H_0:\theta=\theta^\ast,\;–æ–ø—Ä–µ–¥–µ–ª—è—é—â–∞—è\;—Ç–æ—á–Ω–æ–µ\;–∑–Ω–∞—á–µ–Ω–∏–µ\;–ø–∞—Ä–∞–º–µ—Ç—Ä–∞,\;$<br>
$–Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è\;–ø—Ä–æ—Å—Ç–æ–π.$<br>
$–û—Å–Ω–æ–≤–Ω–∞—è\;–≥–∏–ø–æ—Ç–µ–∑–∞:$<br>
$H_0‚à∂\theta\;\in\;\Theta_0$<br>
$–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è\;–≥–∏–ø–æ—Ç–µ–∑–∞:$<br>
$H_A:\theta\in\;\Theta_A:\Theta_0\cap\Theta_A==\;\varnothing$<br>
$–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞\;–∫—Ä–∏—Ç–µ—Ä–∏—è:$<br>
$X^2=\sum_{i=1}^r\frac{\left(n_i-nP_i\right)^2}{nP_i}$<br>

## –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏
–î–ª—è –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –æ–±—Ä–∞–±–æ—Ç–∫—É. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ–¥–µ—Ä–∂–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å.

## –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã

### –ß–∞—Å—Ç—å 1
#### a)–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω—ã–π —Ä—è–¥, —ç–º–ø–∏—Ä–∏—á–µ—Å–∫—É—é —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —á–∞—Å—Ç–æ—Ç


```python
import matplotlib.pyplot as plt
import numpy as np
def graphic_builder(x_values:list,y_values:list,title:str,x_title:str,y_title:str)->None:
    """
    —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
    """
    
    plt.step(x_values,y_values,where="post",color="r")
    plt.xlabel(x_title)
    plt.ylabel(y_title)
    plt.title(title)
    plt.show()

def variation(data:list)->list:
    """
    –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞
    """
    sorted_data:list =sorted(data)
    print(f"–í–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω—ã–π —Ä—è–¥: \n{sorted_data}\n")
    return sorted_data

def empirical_function(sorted_data:list,unique_digits:list)->None:
    """
    –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
    """
   
    #–ø–æ–¥—Å—á–µ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —á–∏—Å–µ–ª –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–∞ –∏–∑ –≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞
    overall_count:int = len(sorted_data)
    digits_dict: dict = {digit: sum(1 for i in data if i==digit) for digit in unique_digits}
    
    #–ß–∞—Å—Ç–æ—Ç–∞ 
    frequency:dict = {digit: digit_count/overall_count for digit,digit_count in digits_dict.items()}
    print(f"–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª: {overall_count}\n")
    for key, value in digits_dict.items():
        print(f"–ß–∏—Å–ª–æ {key} –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ {value}")
    print("\n")
    for key, value in frequency.items():
        print(f"–ß–∏—Å–ª–æ {key} –ß–∞—Å—Ç–æ—Ç–∞ {value}")

    #–ó–Ω–∞—á–µ–Ω–∏–µ –•, –∑–Ω–∞—á–µ–Ω–∏–µ –£
    x_values:list = list(frequency.keys())
    y_values:list = [sum(frequency[digit] for digit in x_values if digit <= x) for x in x_values]
    title:str = "–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è"
    x_title:str = "–ó–Ω–∞—á–µ–Ω–∏—è"
    y_title:str = "–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è —á–∞—Å—Ç–æ—Ç–∞"
    print(f"\n–ó–Ω–∞—á–µ–Ω–∏—è X {x_values}\n")
    print(f"\n–ó–Ω–∞—á–µ–Ω–∏—è Y {y_values}\n")
    x_values.insert(0, min(sorted_data))
    y_values.insert(0, 0)
    
    #–ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
    graphic_builder(x_values,y_values,title,x_title,y_title)
    

def histogram(data:list,unique_digits:list)->None:
    """
    –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã
    """
    plt.hist(sorted_data,bins=(len(unique_digits)), color="white", edgecolor="red",density=True)
    plt.title("–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞")
    plt.xlabel("–ó–Ω–∞—á–µ–Ω–∏—è")
    plt.ylabel("–ß–∞—Å—Ç–æ—Ç–∞")
    plt.show()
    
data:list = [1, 2, 2, 3, 2, 3, 2, 3, 0, 3, 1, 2, 3, 0, 1, 1, 0, 1, 3, 1, 2, 2, 4, 3, 3, 2, 0, 2, 1, 3, 3, 3, 0, 2, 1, 1, 1, 4, 2, 2, 3, 2, 1, 2, 4, 2,3,3,2,2]

#0,1,2,3,4
unique_digits:list = list(set(data))

sorted_data: list = variation(data=data)


```

***–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:***
$$F_n(x)=\left\{\begin{array}{l}0 \;–ø—Ä–∏\;-\infty<x\leq0\\0,1 \;–ø—Ä–∏\;0<x\leq1\\0,32 \;–ø—Ä–∏\;1<x\leq2\\0,66 \;–ø—Ä–∏.\;2<x\leq3\\0,94 \;–ø—Ä–∏\;3<x\leq4\\1 \;–ø—Ä–∏\;4<x\leq+\infty\end{array}\right.$$

#### b)–í—ã—á–∏—Å–ª–∏—Ç—å –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ –∞–Ω–∞–ª–æ–≥–∏ —Å–ª–µ–¥—É—é—â–∏—Ö —á–∏—Å–ª–æ–≤—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫: (i) –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è, (ii)–¥–∏—Å–ø–µ—Ä—Å–∏–∏, (iii) –°–ö–û (iv) –º–µ–¥–∏–∞–Ω—ã, (v)–∞—Å—Å–∏–º–µ—Ç—Ä–∏–∏, (vi) —ç–∫—Å—Ü–µ—Å—Å–∞, (vii)–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏  P(ùëã‚àà[ùëé,ùëè])
  

```python
from scipy.stats import skew,kurtosis
def numerical_characteristics(data:list)->None:
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ –¥–ª—è 2-–≥–æ –∑–∞–¥–∞–Ω–∏—è
    """
    #–ú–∞—Ç.–æ–∂–∏–¥–∞–Ω–∏–µ
    mean = np.mean(data)

    #–î–∏—Å–ø–µ—Ä—Å–∏—è 
    dispersion = np.var(data)
    
    #C–ö–û
    std_deviation = np.std(data)
    
    #–ú–µ–¥–∏–∞–Ω–∞ 
    median = np.median(data)
    
    #–ê—Å—Å–∏–º–µ—Ç—Ä–∏—è 
    assymetric = skew(data)
    
    #–≠–∫—Å—Ü–µ—Å—Å
    excess = kurtosis(data)
    
    #–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è P(X ‚àà [a, b])
    a:float = 0.00
    b: float = 2.85
    possibility:float = sum(1 for digit in data if a<=digit<=b)/len(data)
    print(f"–í—ã–±–æ—Ä–æ—á–Ω—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏:\n"
          f"–ú–∞—Ç.–æ–∂–∏–¥–∞–Ω–∏–µ:{mean},\n"
          f"–î–∏—Å–ø–µ—Ä—Å–∏—è: {dispersion},\n"
          f"–°–ö–û: {std_deviation},\n"
          f"–ú–µ–¥–∏–∞–Ω–∞: {median},\n"
          f"–ê—Å—Å–∏–º–µ—Ç—Ä–∏—è: {assymetric},\n"
          f"–≠–∫—Å—Ü–µ—Å—Å: {excess}\n"
          f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è P(X ‚àà [{a}, {b}]): {possibility}\n")

numerical_characteristics(data = data)
```

#### d)–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ–∫—Å–∏–π –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏  ùõº1 –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞  ùúÜ –Ω–∞ –±–∞–∑–µ –æ—Ü–µ–Ω–∫–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è

$\sqrt n\left(\widehat\theta\left(\overline x\right)-\theta\right)\sim\mathbb{N}\left(0,\frac1{I\left(\theta\right)}\right),\;–≥–¥–µ\;I\left(\theta\right)=\lim_{n\rightarrow\infty}\left(\frac{I_n\left(\theta\right)}n\right)\;–∏\;–≥–¥–µ\;I_n\left(\theta\right)\;-\;–∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è\;–§–∏—à–µ—Ä–∞$<br>

$1,467\leq\lambda\leq2,492$ - –ê.–î.–ò. —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $\alpha_1$ –Ω–∞ –±–∞–∑–µ –û–ú–ü

```python
from scipy.stats import norm
from math import sqrt
def NormPffCalc(a:int)->float:
    """
    –î–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ X(a)
    """
    X_a = norm.ppf(a)
    print(f"–ó–Ω–∞—á–µ–Ω–∏–µ X_a = {X_a}")
    return X_a
    

def Boarders(E: float, n:int,X_a:float):
    return f"–≥—Ä–∞–Ω–∏—Ü–∞: {E-((X_a*sqrt(E))/sqrt(n))}"
    

a1: float=0.01
E: float = np.mean(data)
r:float = 1-a1/2
X_a = NormPffCalc(a=r)
print(f"–õ–µ–≤–∞—è {Boarders(E,n,X_a)}")
print(f"–ü—Ä–∞–≤–∞—è {Boarders(E,n,-X_a)}")
```

```python
from math import exp as exp
from math import factorial as factorial
from scipy.stats import chi2 
def CriteriaCalcWFuncs(m:float,r:int,Criteria:float)->None:
    """
    –ü–æ–¥—Å—á–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—è —Å –ø–æ–º–æ—â—å—é –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π, –∞ —Ç–∞–∫–∂–µ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
    """
    print(
        f"–ó–Ω–∞—á–µ–Ω–∏–µ —Ö–∏^2: {chi2.ppf(m,r)}",
        f"\n–ù–∞–∏–±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏: {1-chi2.cdf(Criteria,r)}"
    )
    

def CriteriaCalc(Pi:list, In:list)->float:
    """
    –ü–æ–¥—Å—á–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—è
    """
    
    squared_diff_sum:float = 0
    Pi[-1]=[sum(Pi[-1])]
    In[-1] = [sum(In[-1])]
    for n_elements, nPi_res in zip(In, Pi):
        for element, nPi in zip(n_elements, nPi_res):
            squared_diff_sum += ((element - nPi)**2) / nPi
            print(f"ni = {element}, nPi = {nPi}:\n(ni-nPi)^2/nPi = {((element - nPi)**2) / nPi}\n")

    print("–ö—Ä–∏—Ç–µ—Ä–∏–π:",squared_diff_sum)
    return squared_diff_sum 

def nPiCalc(lambda_0:float,k:int,n:int)->list[float]:
    """
    –ü–æ–¥—Å—á–µ—Ç nPi –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞
    """
    
    return [(((lambda_0)**k[_]/(factorial(k[_])))*exp(-lambda_0))*n for _ in range(len(k))]


#–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
lambda_0:float =2.00
n: int = 50
a1: float = 0.01
m: float = 1-a1
l_intervals:list=[[0],[1],[2],[3,4]]
n_elements:list = [[5],[11],[17],[14,3]]
r:int = len(l_intervals)-1

nPi_res:list=[nPiCalc(lambda_0,cur_interval,n) for cur_interval in l_intervals]
Criteria:float = CriteriaCalc(nPi_res,n_elements)
CriteriaCalcWFuncs(m,r,Criteria)

```

$\sqrt n\left(\widehat\theta\left(\overline x\right)-\theta\right)\sim\mathbb{N}\left(0,\frac1{I\left(\theta\right)}\right),\;–≥–¥–µ\;I\left(\theta\right)=\lim_{n\rightarrow\infty}\left(\frac{I_n\left(\theta\right)}n\right)\;–∏\;–≥–¥–µ\;I_n\left(\theta\right)\;-\;–∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è\;–§–∏—à–µ—Ä–∞$

```python
from math import exp as exp
from math import factorial as factorial
from scipy.stats import chi2 
def CriteriaCalcWFuncs(m:float,r:int,Criteria:float)->None:
    """
    –ü–æ–¥—Å—á–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—è —Å –ø–æ–º–æ—â—å—é –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π, –∞ —Ç–∞–∫–∂–µ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
    """
    print(
        f"–ó–Ω–∞—á–µ–Ω–∏–µ —Ö–∏^2: {chi2.ppf(m,r)}",
        f"\n–ù–∞–∏–±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏: {1-chi2.cdf(Criteria,r)}"
    )
    

def CriteriaCalc(Pi:list, In:list)->float:
    """
    –ü–æ–¥—Å—á–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—è
    """
    
    squared_diff_sum:float = 0
    Pi[-1]=[sum(Pi[-1])]
    In[-1] = [sum(In[-1])]
    for n_elements, nPi_res in zip(In, Pi):
        for element, nPi in zip(n_elements, nPi_res):
            squared_diff_sum += ((element - nPi)**2) / nPi
            print(f"ni = {element}, nPi = {nPi}:\n(ni-nPi)^2/nPi = {((element - nPi)**2) / nPi}\n")

    print("–ö—Ä–∏—Ç–µ—Ä–∏–π:",squared_diff_sum)
    return squared_diff_sum 

def nPiCalc(lambda_0:float,k:int,n:int)->list[float]:
    """
    –ü–æ–¥—Å—á–µ—Ç nPi –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞
    """
    
    return [((lambda_0)**k[_])/((lambda_0 + 1)**(k[_] + 1)) * n for _ in range(len(k))]



#–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
lambda_0:float =2.00
n: int = 50
a1: float = 0.01
m: float = 1-a1
l_intervals:list=[[0],[1],[2],[3,4]]
n_elements:list = [[5],[11],[17],[14,3]]
r:int = len(l_intervals)-1

nPi_res:list=[nPiCalc(lambda_0,cur_interval,n) for cur_interval in l_intervals]
Criteria:float = CriteriaCalc(nPi_res,n_elements)
CriteriaCalcWFuncs(m,r,Criteria)

```

```python
import numpy as np
from math import exp as exp
from math import factorial as factorial
from scipy.stats import chi2 
def nPiCalc(sigma0:float,alpha0:float,k:float,n:int)->list[float]:
    """
    –ü–æ–¥—Å—á–µ—Ç nPi –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞
    """
    
    return [(((1 / ( np.sqrt(sigma0**1/2*2 * np.pi))) * np.exp(-((cur_k - alpha0) ** 2) / (2 * sigma0))))*n for cur_k in k]

def CriteriaCalcWFuncs(m:float,r:int,Criteria:float)->None:
    """
    –ü–æ–¥—Å—á–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—è —Å –ø–æ–º–æ—â—å—é –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π, –∞ —Ç–∞–∫–∂–µ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
    """
    print(
        f"–ó–Ω–∞—á–µ–Ω–∏–µ —Ö–∏^2: {chi2.ppf(m,r)}",
        f"\n–ù–∞–∏–±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏: {1-chi2.cdf(Criteria,r)}"
    )
    

def CriteriaCalc(Pi:list, In:list)->float:
    """
    –ü–æ–¥—Å—á–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—è
    """
    squared_diff_sum:float = 0
    Pi[0] = [sum(Pi[0])]
    for n_elements, nPi_res in zip(In, Pi):
        for element, nPi in zip(n_elements, nPi_res):
            squared_diff_sum += ((element - nPi)**2) / nPi
            print(f"ni = {element}, nPi = {nPi}:\n(ni-nPi)^2/nPi = {((element - nPi)**2) / nPi}\n")

    print("–ö—Ä–∏—Ç–µ—Ä–∏–π:",squared_diff_sum)
    return squared_diff_sum 



#–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
n: int = 50
a0: float = 35
a2: float = 0.20
sigma0: float = 100
m: float = 1-a2
l_intervals:list=[[-17.15,-13.15,-9.15],[-5.15],[-1.15],[6.25],[10.85],[14.85],[18.85]]
n_elements:list = [[7],[6],[8],[10],[7],[7],[5]]
r:int = len(l_intervals)-1

nPi_res:list=[nPiCalc(sigma0,a0,l_intervals[i],n) for i in range(len(l_intervals))]
Criteria:float = CriteriaCalc(nPi_res,n_elements)
CriteriaCalcWFuncs(m,r,Criteria)
print(f"–ì–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è,—Ç.–∫. 310>8.55")
```

```python
import numpy as np
from scipy.stats import kstest, kstwobign, laplace
from math import sqrt

def KolmogorovCriteria(data: list, cdf: any, alpha2: float) -> None:
    n = len(data)
    ks_statistics, ks_p_value = kstest(data, cdf)

    kvantil = kstwobign.ppf(1 - alpha2)
    temp = sqrt(n) * ks_statistics

    print(f"–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∫—Ä–∏—Ç–µ—Ä–∏—è –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞: {temp}\n",
          f"Ks_p_value: {ks_p_value}\n"
          f"–ö–≤–∞–Ω—Ç–∏–ª—å: {kvantil}\n")

    if temp > kvantil:
        print("–ì–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è")
    else:
        print("–ì–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è")

def Cdf_func(data: list, alpha0: int, sigma0_2: int) -> any:
    sigma0 = sqrt(2) / sqrt(sigma0_2)
    return laplace.cdf(data, loc=alpha0, scale=sigma0)

alpha0 = 35
sigma0_2 = 10
alpha2 = 0.20
print(data)

My_cdf = Cdf_func(data, alpha0, sigma0_2)
KolmogorovCriteria(data, My_cdf, alpha2)
```
